```python
sns.boxplot(y='reading score',data=ds)
```

```python
sns.catplot(y='reading score',kind='violin',data=ds)
```

```python
#distribution plot- represents the overall distribution of continuous data variables
sns.displot(ds['math score'],bins=10)
```

```python
#histogram-represent the data distribution by forming bins along the range of the data 
#          and then drawing bars to show the number of observations that fall in each bin
sns.histplot(ds['math score'],bins=10)
```

```python
sns.histplot(ds['writing score'],bins=15)
```

```python
#scatter plot
#relational plot-allows us to visualise how variables within a dataset relate to each other
sns.relplot(x='writing score',y='reading score',hue='math score',data=ds,palette='rocket')
```

```python
# pairplot-allows us to plot pairwise relationships between variables within a dataset
sns.pairplot(ds)
```

```python
# corelation matrix-statistical technique used to evaluate the relationship between two variables in a data set.
# heatmap- used to plot rectangular data in the form of a color-coded matrix
corelation=ds.corr()
sns.heatmap(corelation,xticklabels=corelation.columns,yticklabels=corelation.columns,annot=True)
```
